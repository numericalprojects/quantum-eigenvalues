```python
# Importing numpy and matplotlib
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
# reading data stored in results_basic.dat

file_name = 'infinite_well_wf.dat'
file_name2 = 'harmonic_oscillator_wf.dat' 
file_name3 = 'woods_saxon_wf.dat'
# loadtxt reads all the data in a file and stores them in an array (assuming it's all floats)
# we use skiprows=1 to avoid the header
all_data = np.loadtxt(file_name,skiprows=1) 
all_data = all_data.transpose()
all_data1 = np.loadtxt(file_name2, skiprows=1) 
all_data1 = all_data1.transpose()
all_data2 = np.loadtxt(file_name3, skiprows=1) 
all_data2 = all_data2.transpose()
# We transpose the array in order to be able to extract each column individually

x_squarewell = all_data[0] 
x_oscillator = all_data1[0] 
x_woods = all_data2[0]
ground_state_square_well = all_data[1]
ground_state_qho = all_data1[1]
ground_state_woods = all_data2[1]
# Each one of these is a numpy array that contains one of the columns in the results_basic.dat file.
# This will allow us to make operations on the extracted data
```

```python
plt.plot(x_squarewell, ground_state_square_well, label = "ground state of infinite square well")
plt.plot(x_oscillator, ground_state_qho, label = "ground state of Quantum Harmonic Oscillator") 
plt.plot(x_woods, ground_state_woods, label = "ground state of Woods Saxon potential")
plt.xlabel('distance x along the potential in meters')
plt.ylabel('Probability density')
# legend displays the labels given in each plot
plt.legend()
plt.show()
```

```python
file_name = 'woods_saxon_ener.dat' 
all_data = np.loadtxt(file_name, skiprows = 1) 
all_data = all_data.transpose() 
radius = all_data[0] 
ground_state_energy = all_data[1] 
first_excited_state_energy = all_data[2] 
second_excited_state_energy = all_data[3] 
```

```python
plt.plot(radius, ground_state_energy, label = "ground state energy of woods saxon")
plt.plot(radius, first_excited_state_energy, label = "First excited state energy of woods saxon") 
plt.plot(radius, second_excited_state_energy, label = "Second excited state energy of woods saxon")
plt.xlabel('distance r along the radius of the woods saxon potential in meters')
plt.ylabel('Energy in MeV')
# legend displays the labels given in each plot
plt.legend()
plt.show()
```
